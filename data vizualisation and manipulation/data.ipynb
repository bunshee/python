```python
import torch as tr
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = tr.arange(-10, 11)
y = x * 2
print(x, y)
```

```python
# Create the figure and axes
fig, ax = plt.subplots()


# Create a line plot
ax.plot(x.numpy(), y.numpy(), marker='o')  # Convert tensors to NumPy arrays

# Add title and labels
ax.set_title('Line Plot of y = x * 2')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Display the plot
plt.show()
```

```python
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
sns.lineplot(x=x, y=y, marker='v')

# Add title and labels
plt.title('Seaborn Line Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

# Display the plot
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tip_by_day = tips.groupby("day")["tip"].mean()
tip_by_day = tip_by_day.reset_index()
tip_by_day
```

```python
sns.barplot(x="tip", y="day",data=tip_by_day)
plt.show
```

```python
sns.pairplot(data=tips, hue="sex")
plt.show()
```

```python
correlation_matrix = tips.corr(numeric_only=True)
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Heatmap of Tips Dataset')

# Display the heatmap
plt.show()
```

```python
amounts_per_day = tips.groupby('day')["total_bill"].sum().reset_index()

sns.barplot(x="day", y="total_bill", data=amounts_per_day)
plt.show()
```
